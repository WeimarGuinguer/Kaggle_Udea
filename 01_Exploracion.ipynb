```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
# Contar el número de estudiantes en cada clase
academic_program = df['ESTU_PRGM_ACADEMICO'].value_counts()

# Mostrar los resultados

print(academic_program)
```

```python
estrato = df.groupby(['ESTU_PRGM_DEPARTAMENTO','FAMI_ESTRATOVIVIENDA']).size().reset_index(name='Cantidad de personas')

estrato
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
estrato.groupby('FAMI_ESTRATOVIVIENDA').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
Enfermos = df[df["ESTU_PRGM_ACADEMICO"] == "ENFERMERIA"]
Enfermos.head()
```

```python
conteo_estudiantes = df.groupby(['ESTU_PRGM_DEPARTAMENTO', 'ESTU_PRGM_ACADEMICO']).size().reset_index(name='Numero de estudiantes')
conteo_estudiantes
```
